# Question 5: Perform data analysis on the poolcnt variable. What do you notice about this variable? How do you think you should handle this if you were to use poolcnt as a variable in a predictive model?

```python
# first take a look at the missing values for this variable
df1[df1.poolcnt.isna()]
```

```python
#it has a lot of missing values for this variable
poolcnt_na_prop = df1[df1.poolcnt.isna()].shape[0]/df1.shape[0]
print(poolcnt_na_prop)
```

```python
#81.9% of the poolcnt has missing value
df1[~df1.poolcnt.isna()].poolcnt.value_counts()
```

### Then I examed the poolcnt which are not missing. The result is there is only one value for poolcnt that is not missing. In this case, it would be better to transform this poolcnt varaible to boolean variable. True for the property has a pool and False for the property that does not has a pool.


```python
df4 = df1
def pool_transform(i):
    if i == 1.0:
        return True
    else :
        return False
```

```python
df4['pool'] = df1.poolcnt.apply(pool_transform)
```

```python
print(df4['pool'].value_counts())
```

# Question 6: Calculate the correlation matrix on the dataset. What observations can you make when looking at this matrix?

```python
corr_matrix = df4.corr()
```

```python
corr_matrix
```

### From this correlation matrix we can see that: 
### 1.the diagonal is always one. Because the correlation of one variable to itself is always one.
### 2. the corrletion matrix is symmetric. Because the correlation variables are cummutative.
### we should focus on either the lower or the upper half of the matrix.
### This matrix has many variables which is hard to see.



```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.heatmap(corr_matrix, cmap="PiYG")
plt.show()
```

```python
corr_pairs = corr_matrix.unstack().sort_values().dropna().sort_values(ascending = False)
```

### Set the threshold for strong correlations as 0.7.
### As the correlation of variable with itself does not provide any insight, filtered out those pairs with ifself.

```python
strong_corr = corr_pairs[(abs(corr_pairs)>0.7) & (abs(corr_pairs)!=1)]
print(strong_corr)
```
