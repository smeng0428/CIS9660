## Question 6: Correlation matrix on dataset

```python
# remove limit on number of columns to show
pd.set_option("display.max_columns", None)
```

```python
# calculate correlation matrix
prop.corr()
```

### There are a lot of variables here, and much to take in.  An initial observation is that there are some missing values in the correlation matrix.  If we investigate these, we can see that the reason why these are occurring is because of constant fields.  Many of the "id" fields contain either a single value or a single non-missing value.  The correlation calculation discards the missing values, so the reason why these are missing has to do with the mathematical formula behind correlation.  If a variable is constant, the denominator of the correlation coefficient formula will be zero, causing a NaN (not a number) value to be returned in Python

### There's a fair amount of correlation between the bedroomcnt and bathroomcnt, which makes sense intuitively

### The tax value fields are highly correlated - ~0.86.  If we're using linear regression to predict home prices, this could cause an issue to include both
